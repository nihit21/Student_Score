```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = df.drop("Unnamed: 0" , axis = 1)
```

```python
sns.countplot(df, x = 'Gender' )
```

```python
df
```

```python
df1 = df.groupby('ParentEduc').agg({'MathScore':'mean' , 'ReadingScore': 'mean' ,'WritingScore' : 'mean' })
print(df1)
```

```python
plt.figure(figsize =(5,5))
sns.heatmap(df1 , annot = True)
```

```python
#from the above chart we have concluded that education of parents have big impact on score
```

```python
df2 = df.groupby('ParentMaritalStatus').agg({'MathScore':'mean' , 'ReadingScore': 'mean' ,'WritingScore' : 'mean'})
```

```python
sns.heatmap(df2, annot = True)
plt.title("maritial status and student score")
```

```python
#from the above chart we can find out that there is no much effect of parental maritial status on scores of the students
```
